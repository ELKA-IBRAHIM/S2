```python
#| label: fig-reg-prop
import numpy as np
from matplotlib import pyplot as plt
from matplotlib_inline import backend_inline
backend_inline.set_matplotlib_formats("svg")

tau = 60e-3

x = np.logspace(-3, 5, 1000)
plt.figure(figsize=(6,5))
Gdb = 20*np.log10((1/3)*np.sqrt(1+(tau*x)**2)/np.sqrt(1+(tau*x/3)**2))
plt.semilogx(x, Gdb, "--")
plt.grid()
plt.xlabel(r"$\omega$ [rad/s]")
plt.ylabel(r"GdB")
plt.show()
```

```python
#| label: fig-reg-prop-integ
import numpy as np
from matplotlib import pyplot as plt
from matplotlib_inline import backend_inline
backend_inline.set_matplotlib_formats("svg")

tau = 50e-3

x = np.logspace(-3, 5, 1000)
plt.figure(figsize=(6,5))
Gdb = 20*np.log10(tau*x/np.sqrt(9+(tau*x)**2))
plt.semilogx(x, Gdb, "--")
plt.grid()
plt.xlabel(r"$\omega$ [rad/s]")
plt.ylabel(r"GdB")
plt.show()
```
